## Section 1. Introduction


### **Interpreter System**
An **interpreter system** is a program that reads and runs instructions in a high-level language directly without first translating them into machine code. As opposed to a compiler, which runs the whole program to executable code before executing it, an interpreter runs and translates commands line by line in real-time. This process facilitates quicker testing, debugging, and instant feedback during program run time, making interpreters critical in most contemporary computing scenarios like scripting, automations, and quick prototyping (Aho et al., 2006; GeeksforGeeks, 2020).


---


### **Concept**
For this project, the team proposes a Face-Shape Aware Makeup Interpreter, a domain-specific interpreter that translates makeup-related commands or styles into step-by-step visual and textual instructions. The system reads user input—either predefined makeup looks ("natural," "glam," or "smokey eye") or specific commands—as well as an uploaded photo of the user's face. It then applies a sequence of actions: face shape detection of the user, creating step-by-step application instructions, and visually overlaying the matching makeup areas over the uploaded photo. By doing so, the interpreter successfully converts user-defined commands into individualized, actionable output, demonstrating how interpreter concepts can be applied beyond typical programming applications.


---


### **Goals**
The primary goal of the designed interpreter is to process structured input—that is, selections from the user's makeup style or text-based commands—and execute these selections or commands to generate dynamic and personalized advice. This entails lexical analysis and parsing of the input, as well as the generation of visual and textual outputs that direct users through every step of the makeup process.

## Section 2. Description of the Input Language


```python
# ============================================================================================
# Section 2. Description of the Input Language
# ============================================================================================

"""
DESCRIPTION OF TOKENS AND GRAMMAR RULES

The DSL uses a command-based syntax with four main command types.
Each command follows specific grammar rules for validation.
"""

# Token Definitions
TOKENS = {
    # Defining reserved keywords ensures predictable lexical behavior in DSL interpreters (Parr, 2010; Finlayson, 2023)
    "COMMANDS": ["/apply", "/style", "/tips", "/blend"],
    # Specifying domain-specific tokens improves semantic clarity and reusability (Kosar et al., 2016)
    "MAKEUP_TYPES": ["foundation", "concealer", "blush", "eyeshadow", "eyeliner",
                     "mascara", "lipstick", "contour", "highlighter"],
    # Defining contextual tokens (face areas) supports rule-based parsing and command validation (Mernik et al., 2011)
    "FACE_AREAS": ["face", "cheeks", "eyelids", "undereyes", "lips", "brows",
                   "nose", "forehead"],
    # Controlled vocabulary for user commands improves natural language alignment (Kosar et al., 2016)
    "STYLES": ["light", "soft", "medium", "bold", "glam", "natural", "smokey"],
    # Standardized color tokens reduce parsing ambiguity and enforce input consistency (Finlayson, 2023)
    "COLORS": ["red", "pink", "brown", "nude", "black", "gold", "silver", "bronze", "coral"],
}


# Grammar Rules in BNF (Backus-Naur Form)
GRAMMAR = """
Grammar Rules (BNF Notation):
=============================

# Using formal grammar notation ensures precise command parsing for domain-specific interpreters (Parr, 2010; Kosar et al., 2016)
<command>        ::= <apply_command> | <style_command> | <tips_command> | <blend_command>

# Structuring each command with clear parameter order enhances readability and usability (Mernik et al., 2011)
<apply_command>  ::= "/apply" <makeup_type> <face_area> <style_or_color>
<style_command>  ::= "/style" <style_name>
<tips_command>   ::= "/tips" <face_area>
<blend_command>  ::= "/blend" <makeup_type>

# Grouping tokens by syntactic roles improves maintainability and supports grammar extensibility (Kosar et al., 2016)
<style_or_color> ::= <style> | <color>
<makeup_type>    ::= foundation | concealer | blush | eyeshadow | eyeliner |
                     mascara | lipstick | contour | highlighter
<face_area>      ::= face | cheeks | eyelids | undereyes | lips | brows | nose | forehead
<style_name>     ::= glam | natural | smokey | light | medium | bold | soft
<color>          ::= red | pink | brown | nude | black | gold | silver | bronze | coral

Command Examples:
=================
# Example commands serve as DSL test cases and documentation for end-user understanding (Raymond, 2023)
/apply foundation face light     → Apply light foundation on face
/style glam                      → Apply complete glam makeup look
/tips cheeks                     → Get tips for cheek makeup
/blend blush                     → Get blending technique for blush
"""

# Section header printing improves interpretability and debugging during interactive execution (Finlayson, 2023)
print("\n" + "=" * 80)
print("SECTION 2: INPUT LANGUAGE SPECIFICATION")
print("=" * 80)
print("\n=== DSL TOKENS ===")

# Iterating through categories for visualization helps validate token definitions and grammar linkage (Raymond, 2023)
for category, tokens in TOKENS.items():
    print(f"\n{category}:")
    print(f"  {', '.join(tokens)}")

# Displaying the entire grammar aids verification and documentation consistency (Kosar et al., 2016)
print("\n" + GRAMMAR)


```

### Description of Tokens and Grammar Rules
Each command in the Makeup Interpreter follows a simple and consistent structure made up of tokens that follow the grammar rules.

1. Command Prefix – Every instruction starts with / followed by a keyword (apply, /style, /tips, /blend).

2. Arguments – After the command, tokens specify what to apply, where, and how.

3. Word Order – The interpreter expects arguments in a fixed order:

    - /apply → makeup_type + face_area + style_or_color

      - Example: /apply foundation face light

    - /style → style_name

        - Example: /style glam

    - /tips → face_area

        - Example: /tips eyes

    - /blend → makeup_type

        - Example: /blend blush

4. Case Insensitivity – Commands and tokens are treated as lowercase for simplicity.

5. Whitespace Separation – Tokens are separated by spaces; no commas or parentheses are used.

### 5.1 Lexer (Tokenizer)
Purpose: Converts raw input string into a stream of validated tokens

Process:
  1. Receives raw command string
  2. Normalizes input (trim, lowercase)
  3. Splits into individual tokens
  4. Validates each token against TOKENS dictionary
  5. Returns clean token list or raises error

Error Handling:
  - Empty input detection
  - Unknown token identification
  - Descriptive error messages

```python
class Lexer:
    def __init__(self, input_text):
        self.input_text = input_text.strip()  # Trims whitespace to prevent lexical ambiguity (Parr, 2010)
        self.tokens = []  # Initializes storage for generated tokens (Kosar et al., 2016)

    def tokenize(self):
        """
        Tokenizes input string and validates against known tokens.

        Returns:
            list: Validated tokens
        Raises:
            ValueError: If input is empty or contains unknown tokens
        """
        if not self.input_text:
            # Ensures non-empty input to avoid null token streams (Mernik et al., 2011)
            raise ValueError("❌ Empty input: Please enter a command")

        # Split and normalize
        self.tokens = self.input_text.lower().split()
        # Normalization simplifies case-insensitive lexical matching for DSLs (Finlayson, 2023).
        # Splitting by whitespace follows common lexical analysis patterns for command interpreters (Kosar et al., 2016).

        # Build valid token set
        valid_tokens = set()
        for token_group in TOKENS.values():
            valid_tokens.update(tok.lower() for tok in token_group)
        # Aggregating known tokens into a unified lookup set allows O(1) validation efficiency (Parr, 2010).

        # Validate each token
        for token in self.tokens:
            if token not in valid_tokens:
                # Raises descriptive errors to support DSL debugging and user guidance (Mernik et al., 2011).
                raise ValueError(
                    f"❌ Lexical Error: Unknown token '{token}'\n"
                    f"Please check available tokens in the command guide"
                )

        return self.tokens  # Returns the validated token stream for parsing (Kosar et al., 2016)

print("\n[5.1] Lexer class implemented ✓")  # Confirms successful class loading during interpreter initialization (Raymond, 2023)
```

### 5.2 Parser

Purpose: Validates token structure and enforces grammar rules

Process:
  1. Receives token list from Lexer
  2. Identifies command type
  3. Validates argument count
  4. Validates argument types against grammar
  5. Returns parse result (success + data structure) or error

Grammar Enforcement:
  - /apply: requires 3 args (makeup_type, face_area, style_or_color)
  - /style: requires 1 arg (style_name)
  - /tips: requires 1 arg (face_area)
  - /blend: requires 1 arg (makeup_type)

Output Format:
  - Success: (True, {"cmd": str, ...parsed_data...})
  - Failure: (False, "error_message")

```python
class Parser:
    def __init__(self, tokens):
        self.tokens = tokens  # Stores tokenized input for grammar validation (Kosar et al., 2016)

    def parse(self):
        """
        Parses token list and validates against grammar rules.

        Returns:
            tuple: (is_valid: bool, result: dict or error_message: str)
        """
        if not self.tokens:
            # Prevents parsing of empty token streams to maintain syntactic integrity (Mernik et al., 2011)
            return False, "❌ Empty token list"

        command = self.tokens[0]  # Extracts primary command keyword for rule dispatch (Parr, 2010)

        # Parse /apply command
        if command == "/apply":
            return self._parse_apply()  # Delegates to rule-specific parsing method (Kosar et al., 2016)

        # Parse /style command
        elif command == "/style":
            return self._parse_style()

        # Parse /tips command
        elif command == "/tips":
            return self._parse_tips()

        # Parse /blend command
        elif command == "/blend":
            return self._parse_blend()

        else:
            # Provides descriptive syntax errors to enhance DSL user experience (Mernik et al., 2011)
            return False, (
                f"❌ Syntax Error: Unknown command '{command}'\n"
                f"Available commands: {', '.join(TOKENS['COMMANDS'])}"
            )

    def _parse_apply(self):
        """Parse /apply command: /apply <makeup> <area> <style_or_color>"""
        if len(self.tokens) != 4:
            # Enforces strict argument count to maintain grammar consistency (Parr, 2010)
            return False, (
                "❌ Syntax Error: /apply requires exactly 3 arguments\n"
                "Usage: /apply <makeup_type> <face_area> <style_or_color>\n"
                "Example: /apply blush cheeks soft"
            )

        _, makeup, area, modifier = self.tokens

        # Validate makeup type
        if makeup not in TOKENS["MAKEUP_TYPES"]:
            # Cross-checks against valid lexical categories to maintain syntactic-semantic alignment (Kosar et al., 2016)
            return False, (
                f"❌ Invalid makeup type '{makeup}'\n"
                f"Available: {', '.join(TOKENS['MAKEUP_TYPES'])}"
            )

        # Validate face area
        if area not in TOKENS["FACE_AREAS"]:
            return False, (
                f"❌ Invalid face area '{area}'\n"
                f"Available: {', '.join(TOKENS['FACE_AREAS'])}"
            )

        # Validate style or color
        if modifier not in TOKENS["STYLES"] + TOKENS["COLORS"]:
            return False, (
                f"❌ Invalid style or color '{modifier}'\n"
                f"Available styles: {', '.join(TOKENS['STYLES'])}\n"
                f"Available colors: {', '.join(TOKENS['COLORS'])}"
            )

        # Returns structured parse tree data for semantic processing (Mernik et al., 2011)
        return True, {
            "cmd": "apply",
            "makeup": makeup,
            "area": area,
            "modifier": modifier
        }

    def _parse_style(self):
        """Parse /style command: /style <style_name>"""
        if len(self.tokens) != 2:
            # Grammar enforcement ensures single-argument rule (Kosar et al., 2016)
            return False, (
                "❌ Syntax Error: /style requires exactly 1 argument\n"
                "Usage: /style <style_name>\n"
                "Example: /style glam"
            )

        style = self.tokens[1]

        if style not in TOKENS["STYLES"]:
            # Style validation maintains DSL consistency with lexical definitions (Finlayson, 2023)
            return False, (
                f"❌ Invalid style '{style}'\n"
                f"Available: {', '.join(TOKENS['STYLES'])}"
            )

        return True, {"cmd": "style", "style": style}

    def _parse_tips(self):
        """Parse /tips command: /tips <face_area>"""
        if len(self.tokens) != 2:
            return False, (
                "❌ Syntax Error: /tips requires exactly 1 argument\n"
                "Usage: /tips <face_area>\n"
                "Example: /tips cheeks"
            )

        area = self.tokens[1]

        if area not in TOKENS["FACE_AREAS"]:
            return False, (
                f"❌ Invalid face area '{area}'\n"
                f"Available: {', '.join(TOKENS['FACE_AREAS'])}"
            )

        return True, {"cmd": "tips", "area": area}

    def _parse_blend(self):
        """Parse /blend command: /blend <makeup_type>"""
        if len(self.tokens) != 2:
            # Requires strict syntax length for predictable parsing behavior (Parr, 2010)
            return False, (
                "❌ Syntax Error: /blend requires exactly 1 argument\n"
                "Usage: /blend <makeup_type>\n"
                "Example: /blend foundation"
            )

        makeup = self.tokens[1]

        if makeup not in TOKENS["MAKEUP_TYPES"]:
            # Ensures semantic accuracy by validating the token category (Kosar et al., 2016)
            return False, (
                f"❌ Invalid makeup type '{makeup}'\n"
                f"Available: {', '.join(TOKENS['MAKEUP_TYPES'])}"
            )

        return True, {"cmd": "blend", "makeup": makeup}

print("[5.2] Parser class implemented ✓")  # Provides execution feedback during interpreter setup (Raymond, 2023)
```

```python
class Executor:
    def __init__(self, parse_result, face_shape=None):
        self.parse_result = parse_result  # Stores parsed command structure for semantic execution (Kosar et al., 2016)
        self.face_shape = face_shape  # Optional contextual variable for adaptive recommendations (Kramer et al., 2020)
        self.face_shape = face_shape.capitalize() if face_shape else None

    def execute(self):
        """Executes the parsed command and generates recommendations."""
        cmd_type = self.parse_result["cmd"]  # Extracts command keyword for dispatching appropriate action (Parr, 2010)

        if cmd_type == "apply":
            return self._execute_apply()
        elif cmd_type == "style":
            return self._execute_style()
        elif cmd_type == "tips":
            return self._execute_tips()
        elif cmd_type == "blend":
            return self._execute_blend()
        # Command dispatching structure follows interpreter design pattern principles (Mernik et al., 2011)

    def _execute_apply(self):
        makeup = self.parse_result["makeup"]
        area = self.parse_result["area"]
        modifier = self.parse_result["modifier"]

        output = [
            f"### 🎨 Applying {makeup.title()}",
            f"**Target Area:** {area.title()}",
            f"**Style/Color:** {modifier.title()}",
            ""
        ]
        # Builds structured Markdown-like output for user readability and UI integration (Norman, 2013)

        if self.face_shape and makeup in FACE_SHAPE_TIPS.get(self.face_shape, {}):
            output.append(f"**✨ Personalized Tip for {self.face_shape.title()} Face:**")
            output.append(FACE_SHAPE_TIPS[self.face_shape][makeup])
            # Injects contextualized advice for personalized output generation (Kramer et al., 2020)

        # Return **both textual output and visualization data** (Kramer et al., 2020)
        return {
            "text": "\n".join(output),  # textual feedback
            "makeup": makeup,
            "area": area,
            "modifier": modifier
        }

    def _execute_style(self):
        """Execute /style command for overall look suggestions"""
        style = self.parse_result["style"]

        output = [
            f"### 💄 Style Recommendation: {style.title()}",
            "",
            "Here’s how to achieve this look:",
        ]
        # Provides structured instructions for user interpretation, enhancing engagement (Norman, 2013)

        style_tips = {
            "glam": [
                "Use bold, dramatic eyeshadows and contouring.",
                "Pair with matte lipstick and highlighter for definition.",
                "Perfect for evening events or photoshoots."
            ],
            "natural": [
                "Opt for nude tones and minimal foundation.",
                "Light mascara and subtle blush bring freshness.",
                "Great for daily wear or casual settings."
            ]
        }
        # Defines a local dictionary for quick style-specific feedback (Kosar et al., 2016)

        if style in style_tips:
            output.extend(f"• {tip}" for tip in style_tips[style])
        else:
            output.append("• No specific tips found for this style.")
            # Gracefully handles missing data for robustness (Raymond, 2023)

        if self.face_shape:
            output.append("")
            output.append(f"**💫 Additional Tip for {self.face_shape.title()} Face:**")
            output.append("Balance features with soft blending and natural tones.")
            # Extends base style recommendations with personalized insight (Kramer et al., 2020)

        return "\n".join(output)

    def _execute_tips(self):
        """Execute /tips command to show advice for a specific area"""
        area = self.parse_result["area"]
        face_shape = self.face_shape

        if not face_shape or face_shape not in FACE_SHAPE_TIPS:
            return f"❌ No tips available for face shape '{face_shape}'."
            # Provides graceful fallback for missing user context (Finlayson, 2023)

        tips_for_shape = FACE_SHAPE_TIPS[face_shape]
        if area not in tips_for_shape:
            return f"❌ No tips found for area '{area}'."
            # Input validation ensures robustness in knowledge retrieval (Kosar et al., 2016)

        tip_text = tips_for_shape[area]
        return f"### 💡 Makeup Tips for {area.capitalize()}\n\n**Your Face Shape:** {face_shape.capitalize()}\n\n{tip_text}"
        # Dynamically constructs the final recommendation string (Norman, 2013)

    def _execute_blend(self):
        makeup = self.parse_result["makeup"]

        output = [
            f"### 🔄 Blending Technique for {makeup.title()}",
            "",
            "**General Technique:**",
            "• Use circular motions for smoother, more natural coverage",
            "• Blend outward from the center for even distribution",
            "• Build up gradually rather than applying heavily at once",
            ""
        ]
        # Structured list format improves cognitive clarity for procedural steps (Norman, 2013)

        if self.face_shape and "general" in FACE_SHAPE_TIPS.get(self.face_shape, {}):
            output.append(f"**For {self.face_shape.title()} Face:**")
            output.append(FACE_SHAPE_TIPS[self.face_shape]["general"])
            # Allows dynamic retrieval of shape-specific blending guidance (Kramer et al., 2020)

        return "\n".join(output)

print("[5.3] Executor class implemented ✓")  # Provides runtime feedback confirming class readiness (Raymond, 2023)
```

## Section 6. Testing with Valid and Invalid Inputs

```python
VALID_TEST_CASES = [
    "/apply foundation face light",
    "/apply blush cheeks soft",
    "/apply eyeshadow eyelids smokey",
    "/style glam",
    "/style natural",
    "/tips cheeks",
    "/tips lips",
    "/blend foundation",
    "/blend blush"
]
# Defines a set of well-formed DSL commands for validation testing (Mernik et al., 2011).
# Positive test cases ensure functional correctness and coverage of primary grammar rules (Kosar et al., 2016).

INVALID_TEST_CASES = [
    ("", "Empty input"),
    ("/paint lips rainbow", "Invalid command and color"),
    ("/apply contour nose sharp", "Invalid style/color"),
    ("/style freestyle", "Invalid style name"),
    ("/apply foundation", "Missing arguments"),
    ("/apply foundation face", "Missing modifier"),
    ("/tips", "Missing face area"),
    ("/blend", "Missing makeup type"),
    ("/apply mascara feet bold", "Invalid face area")
]
# Negative test cases confirm parser resilience against malformed or incomplete input (Finlayson, 2023).
# Including descriptive error expectations supports debugging and regression tracking (Raymond, 2023).

def test_case(input_text, should_fail=False):
    try:
        tokens = Lexer(input_text).tokenize()  # Lexical validation step (Kosar et al., 2016)
        is_valid, result = Parser(tokens).parse()  # Syntax checking through rule-based parsing (Parr, 2010)

        if should_fail:
            if not is_valid:
                print(f"✅ PASS (expected fail) - {input_text}")
                return True
            else:
                print(f"❌ FAIL - Should have failed: {input_text}")
                return False
        else:
            if is_valid:
                print(f"✅ PASS - {input_text}")
                return True
            else:
                print(f"❌ FAIL - {input_text}: {result}")
                return False

    except ValueError as e:
        # Captures lexical errors gracefully for comprehensive testing (Mernik et al., 2011)
        if should_fail:
            print(f"✅ PASS (caught error) - {input_text}")
            return True
        else:
            print(f"❌ FAIL (unexpected error) - {input_text}: {str(e)}")
            return False

def run_tests():
    print("\nTesting valid inputs:")
    valid_passed = sum(test_case(cmd) for cmd in VALID_TEST_CASES)
    # Iterates through valid input samples, measuring parser compliance with grammar rules (Kosar et al., 2016)

    print("\nTesting invalid inputs:")
    invalid_passed = sum(test_case(cmd, should_fail=True) for cmd, _ in INVALID_TEST_CASES)
    # Evaluates robustness against expected failures, essential for fault tolerance validation (Pineau et al., 2021)

    total = len(VALID_TEST_CASES) + len(INVALID_TEST_CASES)
    print(f"\nOverall: {valid_passed + invalid_passed}/{total} tests passed")
    # Produces a quantitative test summary for empirical evaluation of interpreter accuracy (Zhao et al., 2020)

run_tests()  # Automates execution of validation suite to ensure consistent DSL behavior (Raymond, 2023)
```

```python
# SECTION 6: ADVANCED VISUALIZATION WITH LANDMARKS

import mediapipe as mp  # ML-based facial landmark framework for accurate mapping (Lugaresi et al., 2019)
from PIL import Image, ImageDraw  # Pillow for pixel editing and visual overlay rendering (Clark, 2023)
import copy  # Used for undo functionality and image state cloning (Oliphant, 2015)

mp_face = mp.solutions.face_mesh  # Loads MediaPipe FaceMesh model for 468 landmark detection (Bazarevsky et al., 2019)

class Visualizer:
    """
    Interactive visualization for makeup interpreter
    Uses face landmarks for accurate area mapping.
    Supports undo.
    """

    def __init__(self, image_path):
        self.original_image = Image.open(image_path)  # Loads the user image into memory (Clark, 2023)
        self.current_image = self.original_image.copy()  # Creates an editable copy for drawing overlays (Oliphant, 2015)
        self.history = []  # Maintains undo stack for non-destructive edits (Norman, 2013)
        self.landmarks = self._detect_landmarks(self.original_image)  # Detects facial features via ML inference (Lugaresi et al., 2019)
        self.image_size = self.original_image.size  # Stores (width, height) for coordinate scaling (Harris et al., 2020)

    def _detect_landmarks(self, image):
        import cv2
        import numpy as np

        img = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)  # Converts PIL image to OpenCV format (Bradski, 2020)
        with mp_face.FaceMesh(static_image_mode=True) as face_mesh:
            # Static mode for single-image facial analysis (Bazarevsky et al., 2019)
            results = face_mesh.process(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
            if not results.multi_face_landmarks:
                print("⚠️ No face detected")
                return None  # Robust error handling for failed detections (Finlayson, 2023)
            return results.multi_face_landmarks[0].landmark  # Returns list of 468 landmarks (Lugaresi et al., 2019)

    def _get_area_coords(self, area):
        """Return pixel coordinates for given face area using landmarks"""
        if not self.landmarks:
            # fallback to whole image
            w, h = self.image_size
            return [0, 0, w, h]  # Default full-frame region for missing data (Ribeiro et al., 2016)

        w, h = self.image_size
        lm = self.landmarks

        # Landmark indices for basic areas (Mediapipe FaceMesh)
        area_indices = {
            "cheeks": [234, 454],
            "forehead": [10, 338],
            "eyelids": [33, 263],
            "nose": [6, 195],
            "lips": [61, 291]
        }
        # Coordinates correspond to regions of interest based on face topology (Bazarevsky et al., 2019)

        if area not in area_indices:
            return [0, 0, w, h]  # Fallback for invalid regions (Zhao et al., 2019)

        idx1, idx2 = area_indices[area]
        x1 = int(lm[idx1].x * w)
        y1 = int(lm[idx1].y * h)
        x2 = int(lm[idx2].x * w)
        y2 = int(lm[idx2].y * h)
        return [x1, y1, x2, y2]  # Scales normalized coordinates to pixel space (Harris et al., 2020)

    def apply_makeup(self, makeup_type, area, modifier):
        """Apply colored overlay accurately based on real landmark coordinates."""
        if not self.landmarks:
            print("⚠️ No landmarks detected. Makeup not applied.")
            return  # Defensive check to prevent runtime errors (Finlayson, 2023)

        self.history.append(self.current_image.copy())  # Save state for undo feature (Norman, 2013)
        draw = ImageDraw.Draw(self.current_image, "RGBA")  # Enables transparency for realistic overlay (Clark, 2023)

        # Color mapping
        color_map = {
            "blush": (255, 192, 203, 128),
            "contour": (139, 69, 19, 128),
            "highlight": (255, 255, 0, 128),
            "foundation": (245, 222, 179, 128),
            "eyeshadow": (128, 0, 128, 128),
            "lipstick": (255, 0, 0, 128)
        }

        # Default neutral tone fallback (Raymond, 2023)
        color = color_map.get(makeup_type, (128, 128, 128, 128))

        # Modifier-based opacity mapping
        modifier_opacity = {
            "light": 80,
            "soft": 128,
            "medium": 180,
            "bold": 230,
            "glam": 255,
            "natural": 100,
            "smokey": 150
        }
        opacity = modifier_opacity.get(modifier.lower(), 128)  # Default semi-transparent (Raymond, 2023)

        # Apply the modifier to color (preserve RGB, change alpha)
        color = color[:3] + (opacity,)# Preserve RGB, set alpha according to style (Kramer et al., 2020)

        area_map = {
            "cheeks": [[234, 454]],  # left cheek landmark pair
            "forehead": [[10, 338]],
            "eyelids": [[33, 133], [362, 263]],  # left and right eyes
            "nose": [[6, 195]],
            "lips": [[61, 291]]
        }
        # Multi-region mapping improves spatial coverage accuracy (Bazarevsky et al., 2019)

        # scale ellipse size based on modifier intensity
        size_multiplier = {
            "light": 1.0,
            "soft": 1.1,
            "medium": 1.2,
            "bold": 1.3,
            "glam": 1.4,
            "natural": 1.0,
            "smokey": 1.2
        }
        scale = size_multiplier.get(modifier.lower(), 1.0)

        if area not in area_map:
            print(f"⚠️ Unknown area '{area}'. Using whole face instead.")
            w, h = self.current_image.size
            draw.ellipse([0, 0, w, h], fill=color)  # Fallback visualization (Goodfellow et al., 2016)
        else:
            w, h = self.image_size
            for pair in area_map[area]:
                idx1, idx2 = pair
                x1 = int(self.landmarks[idx1].x * w)
                y1 = int(self.landmarks[idx1].y * h)
                x2 = int(self.landmarks[idx2].x * w)
                y2 = int(self.landmarks[idx2].y * h)
                # Expand slightly for visibility
                pad_x = int(0.05 * w)
                pad_y = int(0.05 * h)
                draw.ellipse(
                    [x1 - pad_x*scale, y1 - pad_y*scale, x2 + pad_x*scale, y2 + pad_y*scale],
                    outline=color[:3], width=3, fill=color
                )
                # Visual overlay with transparency mimics digital makeup simulation (Kumar et al., 2021)

        print(f"🖌️ Applied {makeup_type} ({modifier}) on {area}.")
        self.show_image()  # Displays the current result for real-time feedback (Norman, 2013)

    def undo(self):
        """Undo last applied makeup step"""
        if self.history:
            self.current_image = self.history.pop()  # Restores previous saved state (Norman, 2013)
            print("↩️ Undo successful — previous state restored.")
            self.show_image()  # Immediate visual confirmation enhances UX (Kramer et al., 2020)
        else:
            print("⚠️ No previous steps to undo.")  # Feedback for empty history stack (Raymond, 2023)

    def show_image(self):
        """Display current image"""
        display(self.current_image)  # Renders current image inline in notebook (Bisong, 2019)
```

```python
import os  # For handling file system operations (Raymond, 2023)
import glob  # Enables pattern-based file discovery (Oliphant, 2015)

# Automatically detect the first uploaded image file in the directory
supported_formats = ("*.jpg", "*.jpeg", "*.png")  # Common image formats for compatibility (Clark, 2023)
uploaded_files = []
for fmt in supported_formats:
    uploaded_files.extend(glob.glob(fmt))  # Expands wildcard search for image formats (Oliphant, 2015)

if not uploaded_files:
    raise FileNotFoundError("⚠️ No uploaded image file found in the current directory.")
    # Defensive check ensures system robustness for missing uploads (Finlayson, 2023)

# ✅ Sort files by modification time (newest first)
uploaded_files = sorted(uploaded_files, key=lambda x: os.path.getmtime(x), reverse=True)

uploaded_image_path = uploaded_files[0]  # Get the first image file detected (Raymond, 2023)
print(f"✅ Using uploaded image: {uploaded_image_path}")  # Transparent runtime logging (Norman, 2013)

visualizer = Visualizer("cropped_face.jpg")  # Loads visualization model with facial landmarks (Lugaresi et al., 2019)

command_queue = []  # Stores queued makeup application commands for batch visualization (Goodfellow et al., 2016)

while True:
    user_input = input(
        "\nEnter a makeup command, (i.e. /apply, /style, /blend, /tips, 'visualize', 'undo', 'exit'): "
    ).strip()
    # Accepts live user commands with input sanitization (Kosar et al., 2016)

    if user_input.lower() == "exit":
        print("Exiting system.")
        break  # Supports graceful session termination for usability (Norman, 2013)

    elif user_input.lower() == "undo":
        # Removes the last queued makeup command (if any)
        if command_queue:
            removed_command = command_queue.pop()  # Remove the last command entered (Raymond, 2023)
            print(f"↩️ Undid last command: {removed_command['text']}")  # Feedback for user clarity (Norman, 2013)
        else:
            print("⚠️ No commands to undo.")  # Prevents invalid undo operations (Finlayson, 2023)
        continue  # Wait for next user input after undo (Ribeiro et al., 2016)

    elif user_input.lower() == "visualize":
        if not command_queue:
            print("⚠️ No commands to visualize.")  # Inform the user (Ribeiro et al., 2016)
        else:
            for cmd in command_queue:
                visualizer.apply_makeup(cmd["makeup"], cmd["area"], cmd["modifier"])
            visualizer.show_image()  # Display the final styled face (Lugaresi et al., 2019)
            command_queue = []  # Clear after rendering (Raymond, 2023)
        print("✅ Visualization complete. Exiting system.")
        break  # exit runtime after visualization — final stage of interaction (Norman, 2013)

    else:
        try:
            # Detect face shape (placeholder)
            face_shape, _ = detect_face_shape(visualizer.original_image)
            # Hybrid AI integration: CNN for shape + DSL for semantic control (Kramer et al., 2020)

            # Lexer → Parser → Executor
            lexer = Lexer(user_input)  # Token-level validation (Kosar et al., 2016)
            tokens = lexer.tokenize()
            parser = Parser(tokens)  # Grammar validation (Parr, 2010)
            is_valid, result = parser.parse()

            if not is_valid:
                print(result)
                continue  # Allows retry without terminating session (Finlayson, 2023)

            executor = Executor(result, face_shape)  # Binds logic layer with contextual AI (Kramer et al., 2020)
            result = executor.execute()  # Returns dict with 'text', 'makeup', 'area', 'modifier' (Goodfellow et al., 2016)

            print(result)  # Clean Markdown-style output (Norman, 2013)

            # Only queue /apply commands for visualization
            if "makeup" in result:
                command_queue.append(result)  # Adds command to history for later visualization (Norman, 2013)

        except ValueError as e:
            print(e)
            print("Please try again.\n")  # Encourages iterative input correction (Ribeiro et al., 2016)
```

## Section 8. Insights and Conclusions

This project successfully developed an end-to-end makeup recommendation and visualization system that relies on computer vision, natural language processing, and deep learning methods to provide personalized makeup tips for each user. The system demonstrates how different artificial intelligence (AI) fields can be integrated to create an adaptive application capable of understanding human facial structure and user intentions through structured commands.

A MobileNetV2 model was trained on a dataset from Kaggle (Niten19, 2023), which contains labeled images of diverse face shapes, providing a solid foundation for classification. The model proved effective for face shape recognition, achieving medium to high accuracy with minimal computational errors.

Howard et al. (2017) highlighted the strength of MobileNetV2 in mobile and embedded vision applications, confirming that lightweight deep learning models can be optimized for real-time face categorization without sacrificing reliability.

However, it is important to note that the trained model is not perfect and may occasionally misclassify face shapes, particularly under occlusion, unusual angles, or extreme lighting conditions, as this project primarily focuses on creating an interpreter system.

In addition, MediaPipe's Face Mesh was also integrated to improve the accuracy of facial landmark detection. It provided a realistic overlay and precise mapping of cosmetic effects on targeted facial areas on the uploaded image (Lugaresi et al., 2019).

A central feature of the project is the use of a domain-specific language (DSL) combined with the Lexer-Parser-Executor architecture. This allows users to provide structured text-based commands to specify makeup actions (e.g., "/apply blush cheeks soft") for the visualizer. This design simplifies user input and ensures a smooth transition from human interaction to machine execution, which proves how effective a rule-based grammar systems is when it comes to human-computer interfaces (Jurafsky & Martin, 2023). Furthermore, the parser’s robust error-handling mechanisms enhance usability by managing invalid syntax gracefully.

Using color overlays that are in line with facial landmarks, the visualization component accurately replicates the application of makeup from a usability point of view.  AI has the potential to enhance digital cosmetology and self-image evaluation by using visualization as an interpretive link between algorithmic predictions and human perception.  The system's capacity to offer tailored guidance based on identified facial shapes further highlights the growing potential of adaptive AI in the fashion and beauty industries.

In conclusion, this project demonstrates how formal command parsing, computer vision, and machine learning can be integrated to create a cohesive and interactive system for aesthetic personalization.  Future developments might incorporate real-time webcam inference, increase the dataset's generalizability, and use reinforcement learning to gradually improve recommendation accuracy.  Users may occasionally experience misclassifications due to the system's imperfect face shape predictions, which highlights the necessity of human oversight and feedback.  Overall, the system lays a solid basis for upcoming advances in AI-based digital aesthetics, taking into account both useful human-centered applications and technological advancement.
